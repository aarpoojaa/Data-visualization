# Question 1

```python
atom_weight = [1.008,4.003,6.941,9.012,10.811,12.011]
#rounding off to the nearest integer
atom_weight = [round(i) for i in atom_weight]
print(atom_weight)
atomic_elements = ['H', 'He', 'Li', 'Be', 'B', 'C']
print(atomic_elements)
explode1 = [0, 0.1, 0, 0, 0, 0]
explode2 = [0, 0, 0, 0, 0.1, 0]
```

```python
#Pie Chart 1 and 2
fig,ax=plt.subplots(1,2,figsize=(14,6))
ax[0].pie(atom_weight,labels=atomic_elements,explode = explode1,  autopct='%1.1f%%')
ax[0].set_title('Slice annotated with a percentage of the whole')
ax[0].axis('equal')

ax[1].pie(atom_weight,labels=atomic_elements,explode=explode2,  autopct=lambda p: '{:.0f}'.format(p * sum(atom_weight) / 100.0))
ax[1].set_title('Slice annotated with its atomic weight')
ax[1].axis('equal')
plt.show()
```

# Question 3


```python
dates = ['2019-01-01', '2019-01-03', '2019-01-05', '2019-01-07', '2019-01-09', '2019-01-11', '2019-01-13', '2019-01-15']
#Random uniform distribution
dist = np.random.uniform(100,200,8)
dat_frame = pd.DataFrame({'Dates': dates, 'Dist_Value': dist})
#Converting to pandas Datetime objects
dat_frame['Dates'] = pd.to_datetime(dat_frame['Dates'])
dat_frame =  dat_frame.set_index('Dates')

fig, ax = plt.subplots(1, 2, figsize = (15, 10))
f1 = sns.lineplot(x = dat_frame.index, y = dat_frame['Dist_Value'], ax = ax[0])
for item in f1.get_xticklabels():
    item.set_rotation(90)

f2 = sns.barplot(x = dat_frame.index, y = dat_frame['Dist_Value'], ax = ax[1])
for item in f2.get_xticklabels():
    item.set_rotation(90)

ax[0].set_title("Line Chart representiation")
ax[1].set_title("Bar Chart Representation")
plt.tight_layout()
plt.show()
```
