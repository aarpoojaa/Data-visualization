```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Question 1

# Correlation Plot

```python
df = sns.load_dataset('mpg')
df.head()
#Correlation without numeric values plot
f1,ax1=plt.subplots(figsize=(15,10))
sns.heatmap(df.corr(), annot = True, square = True, cmap = 'YlGnBu')
ax1.set_title('Correlation Plot')
plt.tight_layout()
plt.show()
```

From the question, I was confused if the pairplot for the correlation or the pairplot for the full data was asked, so provided both rather than assuming it myself. I didn't put them in the same plot as subplots as one of them is not necessary, so provided them individually. 

# Pairplot of correlation values

```python
sns.set()
f = sns.pairplot(df.corr())
f.fig.suptitle("Pairplot of correlation values", y = 1.01)
plt.show()
```

# Pairplot of the data

```python
sns.set()
f = sns.pairplot(df)
f.fig.suptitle("Pairplot of the data", y = 1.01)
plt.show()
```

# Question 2

```python
df1 = sns.load_dataset('diamonds')
df1.head(10)
```

```python
df1 = df1[df1.color != 'D']
df1 = df1[df1.color != 'E']
df1 = df1[df1.cut != 'Fair']
f2 = sns.FacetGrid(df1, row = 'cut', col = 'color')
f2 = f2.map(plt.scatter, 'price', 'carat')
f2.fig.suptitle("Facetgrid plot", y = 1.03)
#plt.subplots_adjust(top = 0.9)
#f2.fig.suptitle('FACETGRID')
plt.tight_layout()
plt.show()
```

# Question 3

```python
df2 = sns.load_dataset('car_crashes')
fig,axes = plt.subplots(2, 1, figsize = (15,12))
fig.suptitle("Scatter Gram Plots", y = 1.03)
axes[0].set_title("Total vs. Speeding")
sns.regplot('total','speeding', ax = axes[0], data = df2, color = 'red')
axes[1].set_title("Total vs. Alcohol")
sns.regplot('total','alcohol', ax = axes[1], data = df2, color = 'green')
plt.tight_layout()
plt.show()
```

# Question 4

```python
df3 = sns.load_dataset('iris')
fig,axes = plt.subplots(2, 2, figsize = (15,12))
fig.suptitle("Boxplots", y = 1.03)
axes[0,0].set_title('Sepal length distribution')
sns.boxplot(x = 'species', y = 'sepal_length', data = df3, ax = axes[0,0])
axes[0,1].set_title('Sepal width distribution')
sns.boxplot(x = 'species', y = 'sepal_width', data = df3, ax = axes[0,1])
axes[1,0].set_title('Petal length distribution')
sns.boxplot(x = 'species', y = 'petal_length', data = df3, ax = axes[1,0])
axes[1,1].set_title('Petal width distribution')
sns.boxplot(x = 'species', y = 'petal_width', data = df3, ax = axes[1,1])
plt.tight_layout()
plt.show()
```
